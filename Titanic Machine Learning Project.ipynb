```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.20,stratify=target,random_state=42)
```

```python
print(x_train.shape,"\n",x_test.shape,"\n",y_train.shape,"\n",y_test.shape)
```

```python
mms=MinMaxScaler()   ### it scales downs the value between 0 to 1
```

```python
x_train[["age","fare"]]=mms.fit_transform(x_train[["age","fare"]]) ### fit_transform means it learns from the model then transform it
x_test[["age","fare"]]=mms.transform(x_test[["age","fare"]]) ### transform means it only transform the model
```

# logistic Regression 

```python
## it is a classificaton project because in this our dependent value is survived it is binary classification having 0 and 1
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
 log=LogisticRegression()
```

```python
model=log.fit(x_train,y_train)
```

```python
model.score(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_test,pred)
```

```python
cf=confusion_matrix(y_test,pred)
```

```python
cf
```

```python
x_test["actual"]=y_test
x_test["predicted"]=pred
```
